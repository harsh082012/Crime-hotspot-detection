```python
import pandas as pd 
df = pd.read_csv("Delhi_Accident_Data.csv",low_memory=False)
```

```python
df.head()
```

```python
df= df.drop(columns = ['Unnamed: 7','Unnamed: 8','Unnamed: 9'])
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()    #std it describes how far the data points from the mean value if high std means data points are far from mean value and vice versa
"""in 25% it first sort the data points in ascending order and then find the value which is at 25% position similarly for 50% and 75%"""
```

```python
df['DANGER_SCORE'] = (df['# KILLED '] * 3) + (df['# INJURED'])
```

```python
danger_by_district = df.groupby("DISTRICT").agg({      #add injured , killed and danger score for each particukar district
    "# INJURED": "sum",
    "# KILLED ": "sum",
    "DANGER_SCORE": "sum"
}).reset_index()    # add indexes

danger_by_district = danger_by_district.sort_values(by="DANGER_SCORE", ascending=False)

danger_by_district.head(10)
```

```python
danger_by_district.to_csv("delhi_danger_index.csv", index=False)   #index =false Do not write the DataFrameâ€™s index as a column in the CSV file.
```
